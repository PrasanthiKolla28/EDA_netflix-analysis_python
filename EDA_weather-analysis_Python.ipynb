### Find all instances when 'Wind Speed is above 24' and 'Visibility is 25'

### What is the mean value of each column aganist each 'Weather Condition'

```python
numeric_cols = df.select_dtypes(include=['number']).columns
df.groupby('Weather_Condition')[numeric_cols].mean().head(5)
```

### Find all instances when 'Weather is Clear' or 'Visibility is above 40.'

```python
df[(df.Weather_Condition == 'Clear') | (df.Visibility_km >40)]
```

### Find all instances when: 
### a. Weather is 'clear' & 'relative humidity is greater than 50'
### or

```python
df[(df['Weather_Condition'] == 'Clear') & (df['Rel Hum_%'] > 50) | (df['Visibility_km'] > 40)]
```
